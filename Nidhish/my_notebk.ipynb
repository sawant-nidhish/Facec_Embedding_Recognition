```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('../cosineSimMatrix_multiple_imgs.csv')
count_ls=[]
df.set_index('Unnamed: 0',inplace=True)
index=df.index
columns=df.columns

for i in index:
    count=0
    for j in columns:
        if df.loc[i,j] >= 0.2 and i.split("\\")[0]!=j.split("\\")[0] :
           count=count+1
    count_ls.append(count)
count_ls
df['Bad Pairs']=count_ls
df['Bad Pairs']
```

```python
df=pd.read_csv('../cosineSimMatrix_single_imgs.csv')
count_ls1=[]
df.set_index('Unnamed: 0',inplace=True)
index=df.index
columns=df.columns

for i in index:
    count=0
    for j in columns:
        if df.loc[i,j] >= 0.2 and i.split("\\")[0]!=j.split("\\")[0] :
           count=count+1
    count_ls1.append(count)
count_ls1
df['Bad Pairs']=count_ls1
df['Bad Pairs']
```

```python
pd.DataFrame([])
count_ls1=np.array(count_ls1).reshape(-1,1)
count_ls=np.array(count_ls).reshape(-1,1)

pd.DataFrame(data=np.concatenate([count_ls,count_ls1],axis=1),index=df.index,columns=['Multiple_Imgs','Single_imgs'])
```
